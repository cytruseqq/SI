```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
print(tf.__version__)
```

```python
training_data = np.load('sample_data/TF_dataset.npz')
input_size = 2
output_size = 1
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(output_size)
])

model.compile(optimizer='sgd', loss='mean_squared_error')
model.fit(training_data['inputs'], training_data['targets'], epochs=100, verbose=1)
```

```python
model.layers[0].get_weights()
```

```python
plt.plot(
    np.squeeze(model.predict_on_batch(training_data['inputs'])),
    np.squeeze(training_data['targets']),
    color='orange'
)
plt.xlabel('outputs')
plt.ylabel('targets')
plt.show()
```

```python

model_exp = tf.keras.Sequential([
    tf.keras.layers.Dense(output_size)
])

model_exp.compile(optimizer='adam', loss='mean_absolute_error')
model_exp.fit(training_data['inputs'], training_data['targets'], epochs=100, verbose=1)
```

```python
model_nonlinear = tf.keras.Sequential([
    tf.keras.layers.Dense(5, activation='sigmoid'),
    tf.keras.layers.Dense(5, activation='sigmoid'),
    tf.keras.layers.Dense(output_size)
])

model_nonlinear.compile(optimizer='adam', loss='mean_squared_error')
model_nonlinear.fit(training_data['inputs'], training_data['targets'], epochs=100, verbose=1)
```

```python
plt.plot(
    np.squeeze(model_nonlinear.predict_on_batch(training_data['inputs'])),
    np.squeeze(training_data['targets']),
    color='orange'
)
plt.xlabel('outputs')
plt.ylabel('targets')
plt.title('Model nieliniowy (sigmoid)')
plt.show()
```
