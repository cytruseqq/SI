```python
# Blok 1 – import bibliotek
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
print(tf.__version__)
```

```python
# Blok 2 – generowanie danych wejściowych
observations = 1000

xs = np.random.uniform(low=-10, high=10, size=(observations, 1))
zs = np.random.uniform(low=-10, high=10, size=(observations, 1))

inputs = np.column_stack((xs, zs))
print(inputs.shape)
```

```python
# Blok 3 – generowanie targets (celów)
noise = np.random.uniform(low=-1, high=1, size=(observations, 1))

# targets = 2*xs - 3*zs + 5 + noise
# Model powinien nauczyć się wag: w1=2, w2=-3, bias=5
targets = 2 * xs - 3 * zs + 5 + noise

np.savez('TF_dataset', inputs=inputs, targets=targets)
print(targets.shape)
```

```python
# Blok 4 – wizualizacja danych 3D
targets_plot = targets.reshape(observations,)
xs_plot = xs.reshape(observations,)
zs_plot = zs.reshape(observations,)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(xs_plot, zs_plot, targets_plot)
ax.set_xlabel('xs')
ax.set_ylabel('zs')
ax.set_zlabel('Targets')
ax.view_init(azim=100)
plt.show()
```

```python
# Blok 5 – inicjalizacja wag i biasów
init_range = 0.1
weights = np.random.uniform(low=-init_range, high=init_range, size=(2, 1))
biases  = np.random.uniform(low=-init_range, high=init_range, size=1)
print(weights, biases)
```

```python
# Blok 6 – trening modelu (gradient descent, 100 epok)
targets = targets.reshape(observations, 1)
eta = 0.02

for i in range(100):
    outputs = np.dot(inputs, weights) + biases
    deltas  = outputs - targets

    loss = np.sum(deltas ** 2) / 2 / observations
    print(loss)

    deltas_scaled = deltas / observations
    weights = weights - eta * np.dot(inputs.T, deltas_scaled)
    biases  = biases  - eta * np.sum(deltas_scaled)
```

```python
# Blok 7 – sprawdzenie nauczonych wag i biasów
print(weights, biases)
# Oczekiwane: weights ≈ [[2], [-3]], biases ≈ [5]
```

```python
# Blok 8 – porównanie outputs vs targets (dowód na model liniowy)
plt.plot(outputs, targets, color='orange')
plt.xlabel('outputs')
plt.ylabel('targets')
plt.show()
```

## Ćwiczenia

### Ćwiczenie 1 – zmiana liczby próbek
Zmień `observations` na 100, 10 000, 1 000 000 i obserwuj różnice w wynikach.

### Ćwiczenie 2 – zmiana eta
Przetestuj: `eta = 0.0001`, `0.001`, `0.1`, `1`

### Ćwiczenie 3 – nowe targets
Zmień targets na: `targets = 13 * xs + 7 * zs - 12 + noise`  
Model powinien znaleźć: `weights ≈ [[13], [7]]`, `biases ≈ [-12]`

```python
# =============================================
# ĆWICZENIE 3 – nowe targets
# Uruchom ten blok zamiast Bloku 3, a potem
# uruchom ponownie Bloki 5, 6, 7, 8
# =============================================

observations = 1000
xs = np.random.uniform(low=-10, high=10, size=(observations, 1))
zs = np.random.uniform(low=-10, high=10, size=(observations, 1))
inputs = np.column_stack((xs, zs))

noise   = np.random.uniform(low=-1, high=1, size=(observations, 1))
targets = 13 * xs + 7 * zs - 12 + noise

np.savez('TF_dataset_ex3', inputs=inputs, targets=targets)
print('targets shape:', targets.shape)
print('Oczekiwane wagi: w1=13, w2=7, bias=-12')
```

```python
# Ponowna inicjalizacja wag (po zmianie targets)
init_range = 0.1
weights = np.random.uniform(low=-init_range, high=init_range, size=(2, 1))
biases  = np.random.uniform(low=-init_range, high=init_range, size=1)

targets_train = targets.reshape(observations, 1)
eta = 0.02

for i in range(100):
    outputs = np.dot(inputs, weights) + biases
    deltas  = outputs - targets_train
    loss    = np.sum(deltas ** 2) / 2 / observations
    print(f'Epoka {i+1:3d} | Loss: {loss:.4f}')
    deltas_scaled = deltas / observations
    weights = weights - eta * np.dot(inputs.T, deltas_scaled)
    biases  = biases  - eta * np.sum(deltas_scaled)

print('\nNauczone wagi:', weights.T)
print('Nauczony bias:', biases)
print('Oczekiwane:    w1=13, w2=7, bias=-12')
```
